```python
from pymongo import MongoClient
from pprint import pprint

#client = MongoClient("mongodb://admin:secret@localhost:27017/?authSource=admin")
client = MongoClient("mongodb://admin:secret@localhost:27017/?authSource=admin")
```

```python
# Ver bases de datos disponibles
print(client.list_database_names())
```

```python
# Selección de la base de datos
db = client["clase_mongo"]

# Crear la colección (si no existe)
nombre_alumno = "azu"
nombre_coleccion = f"datos_persona_{nombre_alumno}"

# Eliminar si ya existe
if nombre_coleccion in db.list_collection_names():
    db.drop_collection(nombre_coleccion)

db.create_collection(nombre_coleccion)

# Verificar
print(db.list_collection_names())
```

```python
# Verificar
print(db.list_collection_names())
```

```python
# Insertar un documento
collection = db[nombre_coleccion]

collection.insert_one({"nombre": "Juan", "edad": 30})
```

```python
# Consultar todos los documentos de la colección
for doc in collection.find():
    print(doc)
```

```python
# Insertar múltiples documentos
collection.insert_many([
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 35},
    {"nombre": "Pedro", "edad": 18},
    {"nombre": "Carlos", "edad": 40}
])
```

```python
# Consultar documentos con un filtro (edad mayor a 30 años)
for doc in collection.find({"edad": {"$gte": 30}}):
    print(doc)
```

```python
# Obtener un solo documento, si hubiera mas de uno devolverá el primer documento que encuentra en el orden de almacenamiento de la base de datos.
documento = collection.find_one({"nombre": "Juan"})
print(documento)
```

```python
# Actualizar un solo documento
collection.update_one({"nombre": "Juan"}, {"$set": {"edad": 31}})
```

lt = less than, inc = suma 

```python
# Actualizar múltiples documentos
collection.update_many({"edad": {"$lt": 31}}, {"$inc": {"edad": 1}})
```

```python
# Eliminar un solo documento
collection.delete_one({"nombre": "Carlos"})
```

```python
# Eliminar múltiples documentos
collection.delete_many({"edad": {"$gt": 20}})
```
