## üìò 1. Introducci√≥n te√≥rica

Una funci√≥n `lambda` es una funci√≥n an√≥nima, peque√±a y de una sola l√≠nea.

Tiene la forma general:

```python
lambda argumentos: expresi√≥n

Es equivalente a una funci√≥n definida con def, pero m√°s breve y solo para funciones simples.

```markdown
## ‚úÖ 2. ¬øCu√°ndo usar `lambda`?

- Cuando necesitamos funciones simples, de una sola expresi√≥n.
- Para pasarlas como argumentos a funciones como `map`, `filter`, `sorted`, etc.
- Para evitar definir funciones completas con `def` que se usan una sola vez.

### üö´ No se recomienda cuando:
- La funci√≥n es compleja o tiene m√∫ltiples l√≠neas.
- Queremos a√±adir documentaci√≥n o manejo de errores.

````python
# Ejemplo 1: Doble de un n√∫mero
doble = lambda x: x * 2
print(doble(4))
````

````python
# Ejemplo 2: Suma de dos n√∫meros
suma = lambda x, y: x + y
print(suma(3, 5)) 
````

````python
# Ejemplo 3: Obtener el √∫ltimo car√°cter de una palabra
ultimo = lambda s: s[-1]
print(ultimo("funciones lambda en python"))  # n
````

````python
# Ejemplo 4: Ordenar una lista de tuplas por el primer valor
pares = [(1, 3), (4, 1), (2, 2)]
ordenado = sorted(pares, key=lambda x: x[0])
print(ordenado)  

# Ejemplo 4: Ordenar una lista de tuplas por el segundo valor
pares = [(1, 3), (4, 1), (2, 2)]
ordenado = sorted(pares, key=lambda x: x[1])
print(ordenado)  
````

````python
# Ejemplo 5: Filtrar (por modulo) una lista de n√∫meros pares con filter
numeros = [1, 2, 3, 4, 5, 6]
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)  
````

````python
# Ejemplo 6: Elevar al cuadrado con map
numeros = [1, 2, 3, 4]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)  
````

````python
# Ejemplo 7: Verificar si una palabra contiene una letra espec√≠fica
contiene_a = lambda palabra: 'a' in palabra
print(contiene_a("hola"))   
print(contiene_a("hello"))  
````

````python
# Ejemplo 8: Devolver el mayor de tres n√∫meros
mayor = lambda x, y: x if x > y else y
print(mayor(5, 8))  
````

````python
# Ejemplo 9: Ordenar una lista de strings por su longitud
palabras = ["las", "clases", "de", "IBD", "son", "las", "mejores"]
ordenadas = sorted(palabras, key=lambda x: len(x))
print(ordenadas)  # ['es', 'genial', 'python']
````

````python
# Ejemplo 10: Calcular el m√°ximo de dos n√∫meros
mayor = lambda a, b, c: max(a, b, c)
print(mayor(10, 20, 15))  # Salida: 20
````
